## Line Chart
Son utilizados para graficar cualquier tipo de unidades 1D.

```python
x = np.linspace(0, 20, 1000) #Array de 1000 elementos igualmente espaciados entre 0 y 20
```

```python
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y)
```

```python
#Si no estamos en jupyter se usa plt.show() para mostrar el gr√°fico
plt.plot(x,y)
plt.show()
```

### Nombrar los ejes y el gr√°fico

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot'); #para evitar la salida de texto se pone ;
```

### Subplot

```python
#Crear muchos gr√°ficos en uno
plt.subplot(1,2,1) #filas, columnas, posici√≥n
plt.plot(x,y, 'green')

plt.subplot(1,2,2)
plt.plot(y,x, 'red')
```

```python
plt.subplot(2,1,1) #filas, columnas, posici√≥n
plt.plot(x,y, 'green')

plt.subplot(2,1,2)
plt.plot(y,x, 'red')
```

```python
#Implementar gr√°ficos con POO

#1ro crear una figura donde contener el gr√°fico
figura = plt.figure()

#2do crear el gr√°fico a√±adiendo los ejes con sus m√°rgenes y su tama√±o en la figura anterior
grafico = figura.add_axes([0.1,0.1,0.9,0.9]) #m√°rgen izquierdo, m√°rgen superior, tama√±o del gr√°fico horizontal y vertical como % de 1 

#3ro graficar la funci√≥n
#grafico.plot(x,y, 'blue');
```

```python
fig, ax = plt.subplots()
```

```python
#Combinaci√≥n de dos gr√°ficos en la misma figura con distinto tama√±o y distintos m√°rgenes
figura = plt.figure()

grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.17,0.6,0.25,0.25])
grafico2.plot(y,x,'b')
```

```python
#Subplot con POO
figura, graficos = plt.subplots(nrows=2, ncols=2)

#Rellenar cada gr√°fico. Indexaci√≥n como en una matriz
graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

#A√±adir t√≠tulos
graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde');
```

### Tama√±o del gr√°fico

```python
figura = plt.figure(figsize=(8,4)) #tama√±o en pulgadas de la fig

grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r');
```

```python
figura = plt.figure(figsize=(10,2)) #tama√±o en pulgadas de la fig

grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r');
```

### Dos gr√°ficos en la misma figura

```python
figura = plt.figure()

graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='Al cubo')
graficos.legend(loc=0); #loc=0 busca la mejor ubicaci√≥n para poner la leyenda
```

### Tipo de l√≠nea del gr√°fico

```python
#Ancho
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'red',linewidth=5);
```

```python
#Transparencia
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'red',linewidth=5, alpha=0.4);
```

```python
#Tipo de l√≠nea
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'red',linewidth=5, alpha=0.4, linestyle='-.'); #':','-.','-','--'
```

### Marcadores

```python
x = np.linspace(2,9,10)
y = x**2

# Marcar los puntos donde se unen las coordenadas
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'b', marker='o', markersize=10);
```

## Scatterplot
Permite ver los datos desde una perspectiva geom√©trica.


```python
#El 2 es porque el gr√°fico que vamos a hacer esta en 2D
#El 100 puede variar
x = np.random.randn(100, 2)

#Grafico la 1er columna en funci√≥n de la 2da en un gr√°fico 2D
plt.scatter(x[:, 0], x[:, 1]);
```

```python
w = np.random.randint(0,36,200)
z = [x for x in range(200)]

plt.scatter(w,z);
```

```python
#Separaci√≥n de datos por colores

#Datos a visualizar
x = np.random.randn(200, 2)
#Al usar randn tenemos dist normal con los valores centrados en 0.
#Cambiamos el valor central haciendo de la 1er mitad para poder visualizar luego los dos colores separados:
#seleccionar todas las filas de 0 a 50 y a√±adirles 3 a c/u
x[:50] += 3  

#y se usa para asignar distintos valores a los que deben ser de sitintos colores
y = np.zeros(200)
y[:50] = 1 #A la mitad que corrimos le corresponde 1, al resto 0

#Gr√°fico de col 1 vs col 2 como antes pero con colores
plt.scatter(x[:,0], x[:, 1], c=y);
# c tiene que ser 1D array o list que contenga n¬∞ enteros para poder separar los datos
```

## Histogram
Permite ver la distribuci√≥n de probabilidad de los datos.


```python
x = np.random.randn(10000) #Distribuci√≥n normal randn
```

```python
plt.hist(x);
```

```python
#A√±adir m√°s barras logra que el gr√°fico sea m√°s representativo si hay muchos datos
plt.hist(x, bins=50);
```

```python
x = np.random.random(10000); #Dist uniforme random
plt.hist(x, bins=50);
```

```python
#Histograma de 3 col con distintos colores
x = np.random.randn(10000, 3) 
colors = ['yellow', 'blue', 'lime']
plt.hist(x, 30, density=True, color=colors, label=colors);
plt.legend(prop={'size': 10});
plt.title('Histograma comparaci√≥n de columnas en colores');
```

```python
img = Image.open('lena.JFIF')
array = np.array(img)
plt.imshow(array);
plt.title('Lena');
```

```python
#Convertir a blanco y negro:
#Usar la media de las columnas
gray = array.mean(axis=2)
gray.shape
```

```python
plt.imshow(gray, cmap='gray');
plt.title('blanco y negro');
```
