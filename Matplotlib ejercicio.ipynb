Generar y graficar el conjunto de datos correspondiente a una compuerta XOR

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(5000, 2) #Array aleatorio de 5000x2 con dist normal a separar
y = np.zeros(5000) #Array de ceros de 5000x1 para dar color
```

```python
plt.scatter(x[:,0], x[:, 1]); #Gr√°fico ANTES
```

```python
#Determinaci√≥n de condiciones. Cuando x1=x2 poner de un color (y[i]=1), sino de otro color (y[i]=0)

#for i in range(len(y)):
#    if (x[i,0]<=0 and x[i,1]<=0)or(x[i,0]>=0 and x[i,1]>=0):
#        y[i]=1   

for i in range(len(y)):
    if x[i,0]==x[i,1]:
        y[i]=1   

plt.scatter(x[:,0], x[:, 1], c=y); #Gr√°fico SEPARADO
```

```python
#Gr√°fico DESPU√âS
plt.scatter(x[:,0], x[:, 1], c=y);
plt.xlim(-1, 1);
plt.ylim(-1, 1);
```
